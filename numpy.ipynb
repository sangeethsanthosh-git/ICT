```python
import numpy as np
import pandas as pd

df = pd.read_csv("file.csv")

df.columns = df.columns.str.strip().str.replace(" ", "_")


mpg = df["mpg"].to_numpy()

mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

count_above_25 = np.sum(mpg > 25)

print("Mean MPG:", mean_mpg)
print("Median MPG:", median_mpg)
print("Std MPG:", std_mpg)
print("Cars with MPG > 25:", count_above_25)


cyl = df["cylinders"].to_numpy()
car_names = df["car_name"].to_numpy()

filtered_cars = car_names[cyl > 6]

print("\nCars with >6 cylinders:")
print(filtered_cars.tolist())

hp = pd.to_numeric(df["horsepower"], errors="coerce").to_numpy()

hp_mean = np.nanmean(hp)
hp_clean = np.where(np.isnan(hp), hp_mean, hp)

hp_increased = hp_clean * 1.10

print("\nHorsepower increased by 10%:")
print(hp_increased[:10])


origin = df["origin"].to_numpy()
displacement = df["displacement"].to_numpy()

europe_disp = displacement[origin == 2]
avg_europe_disp = np.mean(europe_disp)

print("\nAverage displacement (Europe):", avg_europe_disp)


unique_cyl = np.unique(cyl)

print("\nMean MPG by cylinders:")
for c in unique_cyl:
    mean_val = np.mean(mpg[cyl == c])
    print(f"{c} cylinders → {mean_val}")
```
