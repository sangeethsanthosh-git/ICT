```python
import pandas as pd
import numpy as np


df = pd.read_csv("file.csv")

df.columns = df.columns.str.strip().str.replace(" ", "_")

print("Dataset Loaded Successfully\n")

print("---- First 10 Rows ----")
print(df.head(10))

print("\n---- Shape ----")
print("Rows, Columns:", df.shape)

print("\n---- Summary Statistics ----")
print(df.describe())


print("\n---- Cars from 1975 & weight < 3000 ----")
filtered = df[(df["model_year"] == 75) & (df["weight"] < 3000)]
result = filtered[["name", "weight", "mpg"]]
print(result)


print("\n---- Missing Values ----")
print(df.isnull().sum())

median_hp = df["horsepower"].median()
df["horsepower"].fillna(median_hp, inplace=True)

print("\nMissing horsepower replaced with median")


df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]

print("\n---- Power to Weight Ratio ----")
print(df[["horsepower", "weight", "power_to_weight_ratio"]].head())


print("\n---- Mean MPG by Origin ----")
grouped = df.groupby("origin")["mpg"].mean()
print(grouped)


print("\n---- Summary by Model Year ----")
summary_df = df.groupby("model_year")[["mpg", "weight", "horsepower"]].mean()
print(summary_df)


high_mpg = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]

high_mpg.to_csv("high_mpg_cars.csv", index=False)

print("\nFile saved as high_mpg_cars.csv")
```
