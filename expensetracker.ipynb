Expense Tracker

```python

FILE_NAME = "expenses.txt"

def add_expense():
    try:
        category = input("Enter category (Food/Travel/etc): ").strip()
        amount = float(input("Enter amount: "))  # may raise ValueError

        with open(FILE_NAME, "a") as file:
            file.write(f"{category},{amount}\n")

        print("Expense added successfully!\n")

    except ValueError:
        print("Invalid amount! Please enter a number.\n")

def view_expenses():
    try:
        with open(FILE_NAME, "r") as file:
            lines = file.readlines()

            if not lines:
                print("No expenses found.\n")
                return

            print("\n--- All Expenses ---")
            for line in lines:
                category, amount = line.strip().split(",")
                print(f"Category: {category} | Amount: {amount}")
            print()

    except FileNotFoundError:
        print("No expense file found. Add expenses first.\n")

def calculate_total():
    try:
        total = 0
        with open(FILE_NAME, "r") as file:
            for line in file:
                _, amount = line.strip().split(",")
                total += float(amount)

        print(f"\nTotal Expense: {total}\n")

    except FileNotFoundError:
        print("No expense file found. Add expenses first.\n")

def category_count():
    try:
        counts = {}

        with open(FILE_NAME, "r") as file:
            for line in file:
                category, _ = line.strip().split(",")
                counts[category] = counts.get(category, 0) + 1

        print("\n--- Category Count ---")
        for cat, count in counts.items():
            print(f"{cat}: {count}")
        print()

    except FileNotFoundError:
        print("No expense file found. Add expenses first.\n")

def main():
    while True:
        print("===== Expense Tracker =====")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Total Expense")
        print("4. Category Count")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            calculate_total()
        elif choice == "4":
            category_count()
        elif choice == "5":
            print("Exiting program...")
            break
        else:
            print("Invalid choice! Try again.\n")

main()
```
