```python
import matplotlib.pyplot as plt
import numpy as np
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
prices = [450, 470, 490, 510, 495, 505, 520]
tar_price = 500
plt.plot(days, prices, marker='o', label="Stock Price")
plt.axhline(y=tar_price, color='r', linestyle='--', label="Target = 500")
plt.axhline(y=500, color="red", linestyle="--")
above_target = [p if p > tar_price else np.nan for p in prices]
plt.scatter(days, above_target, label="Above Target")
plt.axhline(y=500, color="red", linestyle="--", label="Target = 500")
plt.axhline(y=max(prices), color="green", linestyle="--", label="Max Price")
plt.axvline(x=max(days, key=lambda d: abs(tar_price - prices[days.index(d)])), color="green", linestyle="--", label="Max Price Day")
plt.title("7 days stock alert ")
plt.xlabel("days")
plt.ylabel("price")
plt.legend()
plt.grid()
plt.show()
```
